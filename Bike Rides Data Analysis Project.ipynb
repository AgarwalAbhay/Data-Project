```python
# Creating the dictionary to sepcify the coloumn names to be used - 
new_col_names_dict = {'timestamp':'time', 
                         'cnt':'count', 
                         't1':'real_temp', 
                         't2':'temp_feel_like', 
                         'hum':'humidity_percent', 
                         'wind_speed':'wind_speed_kmph',
                         'weather_code': 'weather',
                         'is_holiday': 'is_holiday',
                         'is_weekend': 'is_weekend',
                         'season': 'season'
                     }
```

```python
# Renaming the columns to the required names
df.rename(columns = new_col_names_dict, inplace = True)
df
```

```python
# Chaning the humidity values to percentage
df['humidity_percent'] = df['humidity_percent']/100
df
```

```python
# Creating a weather dictionary to map the written codes to actual values
weather_dict = {'1.0' : 'Clear',
                '2.0' : 'Scattered Clouds',
                '3.0' : 'Broken Clouds',
                '4.0' : 'Cloudy',
                '7.0' : 'Rain',
                '10.0' : 'Rain with Thunderstorm',
                '26.0' : 'Snowfall'}

# Replacing the existing values with the required weather details
df['weather'] = df['weather'].astype('str')
df['weather'] = df['weather'].map(weather_dict)
df
```

```python
# Creating a season dictionary to map the written codes to actual values
season_dict = {'1.0' : 'Summer',
                '2.0' : 'Autumn',
                '3.0' : 'Winter',
               '0.0' : 'Spring'}

# Replacing the existing values with the required weather details
df['season'] = df['season'].astype('str')
df['season'] = df['season'].map(season_dict)
df
```

```python
# Extracting the final dataframe to an excel
df.to_excel('london_bikes_final.xlsx', sheet_name = 'Data')
```
